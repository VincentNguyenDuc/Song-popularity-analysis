```python
import os
import sys

current_path = os.getcwd()
data_path = current_path.replace('src', 'data')
```

```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 7.5]
```

```python
# import data
song_data = pd.DataFrame(pd.read_csv(data_path + '\\raw\\song_data.csv'))
song_data
```

```python
# choosing features
song_data.corr()
```

```python
# divide the data set randomly into a 75% training set and a 25% testing set
features = song_data.iloc[:, 2:]
popularity = song_data.iloc[:, 1]
features_train, features_test, popularity_train, popularity_test = train_test_split(features, popularity, test_size=0.25, random_state=0)
```

```python
# write testing data to csv
pd.DataFrame(features_test).to_csv(data_path + '\\testing\\features_test.csv', index=False)
pd.DataFrame(popularity_test).to_csv(data_path + '\\testing\\popularity_test.csv', index=False)
```

```python
# write training data to csv
pd.DataFrame(features_train).to_csv(data_path + '\\training\\features_train.csv', index=False)
pd.DataFrame(popularity_train).to_csv(data_path + '\\training\\popularity_train.csv', index=False)
```
